#Split dataset



```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('species', axis=1), data['species'], test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

##Model Selection

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
```

```python
dt = DecisionTreeClassifier()
lr = LogisticRegression()
svm = svm.SVC()
knn = KNeighborsClassifier()
```

##Training on without scaled data

```python
dt.fit(X_train, y_train)
lr.fit(X_train, y_train)
svm.fit(X_train, y_train)
knn.fit(X_train, y_train)
```

```python
dt_prediction = dt.predict(X_test)
lr_prediction = lr.predict(X_test)
svm_prediction = svm.predict(X_test)
knn_prediction = knn.predict(X_test)
```

```python
print('accuracy of Decision Tree :',accuracy_score(y_test, dt_prediction))
print('accuracy of Logistic Regression :',accuracy_score(y_test, lr_prediction))
print('accuracy of SVM :',accuracy_score(y_test, svm_prediction))
print('accuracy of KNN :',accuracy_score(y_test, knn_prediction))
```

```python
a_index = list(range(1, 11))
accuracies = []

for i in a_index:
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    accuracies.append(accuracy_score(y_test, prediction))

a = pd.Series(accuracies, index=a_index)

plt.plot(a_index, accuracies)
plt.xticks(a_index)
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy vs. Number of Neighbors')
plt.grid(True)
plt.show()
```

##Training on scaled data

```python
dt.fit(X_train_scaled, y_train)
lr.fit(X_train_scaled, y_train)
svm.fit(X_train_scaled, y_train)
knn.fit(X_train_scaled, y_train)
```

```python
dt_prediction = dt.predict(X_test_scaled)
lr_prediction = lr.predict(X_test_scaled)
svm_prediction = svm.predict(X_test_scaled)
knn_prediction = knn.predict(X_test_scaled)
```

```python
print('accuracy of Decision Tree :',accuracy_score(y_test, dt_prediction))
print('accuracy of Logistic Regression :',accuracy_score(y_test, lr_prediction))
print('accuracy of SVM :',accuracy_score(y_test, svm_prediction))
print('accuracy of KNN :',accuracy_score(y_test, knn_prediction))
```

##Voting

```python
dt = DecisionTreeClassifier()
lr = LogisticRegression(max_iter=1000)
knn = KNeighborsClassifier()
```

```python
estimators = [('dt', dt), ('lr', lr), ('svm', svm), ('knn', knn)]
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
print(X.shape)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
vc = VotingClassifier(estimators = estimators, voting='hard')
x = cross_val_score(vc, X, y, cv=10, scoring='accuracy')
print(np.round(np.mean(x),2))
```

```python
vc = VotingClassifier(estimators = estimators, voting='hard')
x = cross_val_score(vc, X_scaled, y, cv=10, scoring='accuracy')
print(np.round(np.mean(x),2))
```
